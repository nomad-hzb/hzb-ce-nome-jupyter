# Owned Uploads

The following list shows all uploads where the current user is the main author.

```python
import requests
import os
import pandas as pd
from IPython.display import display, HTML

url = "https://nomad-hzb-ce.de/nomad-oasis/api/v1"
token = os.environ['NOMAD_CLIENT_ACCESS_TOKEN']
```

```python
def get_all_owned_uploads(url, token, number_of_uploads=20):
    response = requests.get(f'{url}/uploads',
                             headers={'Authorization': f'Bearer {token}'},params=dict(page_size=number_of_uploads,order_by='upload_create_time', order="desc", roles="main_author"))
    return response.json()["data"]
```

```python
uploads = get_all_owned_uploads(url, token, number_of_uploads=400)
df = pd.DataFrame(uploads)
df['upload_id & NOMAD link'] =  df.apply(lambda x: f'<a href="https://nomad-hzb-ce.de/nomad-oasis/gui/user/uploads/upload/id/{x['upload_id']}" target="_blank">{x['upload_id']}</a>', axis=1)
display(HTML(df[['upload_create_time', 'upload_name', 'upload_id & NOMAD link', 'entries']].to_html(escape=False)))
```
